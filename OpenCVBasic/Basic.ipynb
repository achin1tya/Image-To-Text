```python
img1 = cv2.imread('3D-Matplotlib.png',cv2.IMREAD_COLOR)
img2 = cv2.imread('mainsvmimage.png',1)
```

```python
add1 = img1 + img2

add2 = cv2.add(img1, img2)

weighted = cv2.addWeighted(img1, 0, img2, 0, 0)
```

```python
img2gray = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

# 220 is Threshold value...if pixel > 220 that will become 255 and pixel < 220 will become 0
ret, mask = cv2.threshold(img2gray, 220, 255, cv2.THRESH_BINARY_INV)
ret, mask1 = cv2.threshold(img2gray, 220, 255, cv2.THRESH_BINARY)
```

```python
cv2.imshow('thresh', mask1)
cv2.imshow('thresh_inv', mask)
cv2.imshow('org', img2gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
mask_inv = cv2.bitwise_not(mask)
```

```python
cv2.imshow('org', mask_inv)
cv2.imshow('thresh_inv', mask)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('bookpage.jpg', 1)
grayscaled = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
ret, thresh = cv2.threshold(image, 12, 255, cv2.THRESH_BINARY)
cv2.imshow('thr',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
th = cv2.adaptiveThreshold(grayscaled, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 1)
cv2.imshow('original',image)
cv2.imshow('Adaptive threshold',th)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
